## Background: What is `np.random`?

`np` is the usual nickname for NumPy:

```python
import numpy as np
```

`np.random` is NumPy's randomness module. In Assignment 2, we use it to simulate outcomes that are uncertain.

Useful functions:

- `np.random.choice(options, p=probabilities)` chooses one item from a list of options.
- `np.random.uniform()` returns a random number between 0 and 1.
- `np.random.seed(...)` makes random results repeatable while debugging.

````python
import numpy as np

np.random.seed(1)

# Choose between two labels with equal probability.
print(np.random.choice(["A wins", "B wins"]))

# Choose between 0 and 1, where 1 has probability 0.7.
print(np.random.choice([0, 1], p=[0.3, 0.7]))

# Draw a random grade between 0 and 1.
print(np.random.uniform())
````
