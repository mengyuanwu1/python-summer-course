## Section 4: Dictionaries

Dictionaries store key-value pairs. They are good for labeled information.

```python
student = {
    "name": "Lina",
    "grade_level": 11,
    "favorite_subject": "chemistry",
}
student["grade_level"] = 12
for key, value in student.items():
    print(key, "->", value)
```

## Section 5: Lists of Dictionaries

A list of dictionaries is a simple way to represent rows of tabular data.

```python
students = [
    {"name": "Ava", "major_interest": "biology", "scores": [90, 95]},
    {"name": "Ben", "major_interest": "business", "scores": [82, 78]},
    {"name": "Carmen", "major_interest": "biology", "scores": [96, 94]},
]

for student in students:
    average = sum(student["scores"]) / len(student["scores"])
    print(student["name"], average)
```

### Exercise 5.1: Filter Student Records and Count Interests

```python
records = [
    {"name": "Ava", "interest": "biology", "average": 92},
    {"name": "Ben", "interest": "business", "average": 84},
    {"name": "Carmen", "interest": "biology", "average": 95},
    {"name": "Dev", "interest": "computer science", "average": 89},
]

high_averages = []
interest_counts = {}
for record in records:
    if record["average"] > 90:
        high_averages.append(record["name"])
    interest = record["interest"]
    interest_counts[interest] = interest_counts.get(interest, 0) + 1
print(high_averages)
print(interest_counts)
```

**TA Note:** Expected high averages: Ava and Carmen.

**Hint to give:** Ask students to solve filtering first, then counting.

**Common student mistakes:** Forgetting to initialize a count for new interests.

## Section 6: Object-Oriented Programming

A class is a blueprint. An object is one example made from that blueprint.

```python
class Student:
    def __init__(self, name, grade_level, scores):
        self.name = name
        self.grade_level = grade_level
        self.scores = scores

    def average_score(self):
        return sum(self.scores) / len(self.scores)

    def is_passing(self):
        return self.average_score() >= 60

student = Student("Nora", 11, [88, 91, 85])
print(student.name)
print(student.average_score())
print(student.is_passing())
```

### Exercise 6.1: Build a Movie Class

```python
class Movie:
    def __init__(self, title, genre, rating):
        self.title = title
        self.genre = genre
        self.rating = rating

    def recommendation(self):
        return f"If you like {self.genre}, try {self.title}. It has a rating of {self.rating}."

movie = Movie("Hidden Figures", "drama", 8.0)
print(movie.recommendation())
```

**TA Note:** Any clear recommendation sentence is acceptable.

**Hint to give:** Ask: what data does each movie object need to remember?

**Common student mistakes:** Forgetting self, or using local variables instead of attributes.

## Section 7: Mini Project

Build a simple gradebook:

- Store multiple students.
- Add scores.
- Calculate each student's average.
- Print a class summary.

```python
gradebook = [
    Student("Ava", 10, [90, 92, 88]),
    Student("Ben", 10, [78, 81, 84]),
    Student("Carmen", 10, [95, 97, 94]),
]

class_total = 0
for student in gradebook:
    avg = student.average_score()
    class_total += avg
    print(f"{student.name}: {avg:.1f} ({'passing' if student.is_passing() else 'not passing'})")

print(f"Class average: {class_total / len(gradebook):.1f}")
```

**TA Note:** Expected class average with the example is about 89.0.

**Hint to give:** Encourage students to reuse the Student class.

**Common student mistakes:** Rewriting average logic in many places instead of using the method.
