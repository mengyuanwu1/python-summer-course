## Section 4: Loops

Loops repeat code. A `for` loop is useful when you know what collection to repeat over.

```python
total = 0
for number in range(1, 11):
    total += number
print(total)

for number in range(1, 21):
    if number % 2 == 0:
        print(number)

scores = [75, 82, 91, 68, 88]
count_above_80 = 0
for score in scores:
    if score > 80:
        count_above_80 += 1
print("Scores above 80:", count_above_80)
```
