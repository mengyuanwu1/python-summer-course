## Section 3: Conditionals

Conditionals let a program choose different actions.

```python
score = 87

if score >= 90:
    letter = "A"
elif score >= 80:
    letter = "B"
elif score >= 70:
    letter = "C"
else:
    letter = "Needs more practice"

print(letter)

age = 12
if age < 13:
    print("Child ticket")
elif age < 18:
    print("Teen ticket")
else:
    print("Adult ticket")
```

## Section 4: Loops

Loops repeat code. A `for` loop is useful when you know what collection to repeat over.

```python
total = 0
for number in range(1, 11):
    total += number
print(total)

for number in range(1, 21):
    if number % 2 == 0:
        print(number)

scores = [75, 82, 91, 68, 88]
count_above_80 = 0
for score in scores:
    if score > 80:
        count_above_80 += 1
print("Scores above 80:", count_above_80)
```

## Section 5: Functions

A function is a reusable block of code. `return` gives a value back to the caller. `print` only displays a value.

```python
def calculate_average(scores):
    return sum(scores) / len(scores)

calculate_average([90, 85, 100])
```

### Exercise 5.1: Celsius to Fahrenheit

```python
def celsius_to_fahrenheit(celsius):
    return celsius * 9 / 5 + 32

print(celsius_to_fahrenheit(0))
print(celsius_to_fahrenheit(100))
```

### Exercise 5.2: Even Number Checker

```python
def is_even(number):
    return number % 2 == 0

print(is_even(10))
print(is_even(7))
```

### Exercise 5.3: Letter Grade Function

```python
def letter_grade(score):
    if score >= 90:
        return "A"
    elif score >= 80:
        return "B"
    elif score >= 70:
        return "C"
    elif score >= 60:
        return "D"
    else:
        return "F"

print(letter_grade(92))
print(letter_grade(76))
```

**TA Note:** Expected outputs include 32.0, 212.0, True, False, A, and C.

**Hint to give:** Ask students to describe the formula or grade rule in words before coding.

**Common student mistakes:** Printing instead of returning, or putting broad conditions before narrow ones.

## Section 7: Challenge

Build a grade calculator.

Input: a list of assignment scores.

Output: average score and letter grade.

```python
def letter_grade(score):
    if score >= 90:
        return "A"
    elif score >= 80:
        return "B"
    elif score >= 70:
        return "C"
    elif score >= 60:
        return "D"
    else:
        return "F"

def grade_calculator(scores):
    average = sum(scores) / len(scores)
    grade = letter_grade(average)
    return average, grade

average, grade = grade_calculator([88, 92, 95, 81])
print(f"Average: {average:.1f}")
print(f"Letter grade: {grade}")
```

**TA Note:** Expected output: Average 89.0 and Letter grade B with the scale above.

**Hint to give:** Point students back to calculate_average and letter_grade.

**Common student mistakes:** Forgetting to call the helper function, or returning only one value.
