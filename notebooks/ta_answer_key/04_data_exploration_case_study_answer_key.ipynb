## Section 2: Import Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

## Section 3: Load a Dataset

```python
from pathlib import Path

DATA_DIR = Path("../../data")
if not (DATA_DIR / "sample_students.csv").exists():
    DATA_DIR = Path("data")

csv_path = DATA_DIR / "sample_students.csv"
print("Using:", csv_path)
```

```python
df = pd.read_csv(csv_path)
df.head()
```

```python
print("Shape:", df.shape)
print("Columns:", list(df.columns))
df.info()
```

### Exercise 3.1: Identify Rows, Columns, Numeric Columns, and Categorical Columns

```python
row_count, column_count = df.shape
numeric_columns = df.select_dtypes(include="number").columns.tolist()
categorical_columns = df.select_dtypes(exclude="number").columns.tolist()
print(row_count, column_count)
print(numeric_columns)
print(categorical_columns)
```

**TA Note:** Expected shape is 15 rows by 7 columns.

**Hint to give:** Ask students what `.shape` returns.

**Common student mistakes:** Mixing up rows and columns, or treating IDs as meaningful measurements.

## Section 4: Inspect and Clean Data

Cleaning includes missing values, duplicates, filtering, and creating useful new columns.

```python
print(df.isna().sum())
print("Duplicate rows:", df.duplicated().sum())

df_clean = df.copy()
df_clean["english_score"] = df_clean["english_score"].fillna(df_clean["english_score"].mean())
df_clean["average_score"] = df_clean[["math_score", "english_score", "science_score"]].mean(axis=1)
df_clean.head()
```

### Exercise 4.1: Missing Values, Filtering, and a New Column

```python
missing_by_column = df.isna().sum()
high_math = df[df["math_score"] >= 90]
df_exercise = df.copy()
df_exercise["science_minus_math"] = df_exercise["science_score"] - df_exercise["math_score"]
print(missing_by_column)
print(high_math[["name", "math_score"]])
print(df_exercise[["name", "science_minus_math"]].head())
```

**TA Note:** One English score is missing in the toy data.

**Hint to give:** Ask students what kind of result `df.isna().sum()` returns.

**Common student mistakes:** Changing the original df by accident before comparing cleaned and raw data.

## Section 5: Exploratory Data Analysis

Use summaries to understand patterns before modeling.

```python
print(df_clean.describe())
print(df_clean["favorite_subject"].value_counts())
print(df_clean.groupby("grade_level")["average_score"].mean())
```

### Exercise 5.1: Average by Group and Highest/Lowest Values

```python
average_by_subject = df_clean.groupby("favorite_subject")["average_score"].mean()
highest_student = df_clean.loc[df_clean["average_score"].idxmax(), ["name", "average_score"]]
lowest_student = df_clean.loc[df_clean["average_score"].idxmin(), ["name", "average_score"]]
print(average_by_subject)
print(highest_student)
print(lowest_student)
print("Pattern: In this small dataset, favorite-subject groups have different average scores, but the sample is too small for a strong claim.")
```

## Section 6: Visualization

Use clear titles and axis labels. A plot without labels is hard to interpret.

```python
plt.figure(figsize=(7, 4))
plt.plot(df_clean["student_id"], df_clean["average_score"], marker="o")
plt.title("Average Score by Student ID")
plt.xlabel("Student ID")
plt.ylabel("Average Score")
plt.show()

plt.figure(figsize=(7, 4))
df_clean.groupby("favorite_subject")["average_score"].mean().plot(kind="bar")
plt.title("Average Score by Favorite Subject")
plt.xlabel("Favorite Subject")
plt.ylabel("Average Score")
plt.show()

plt.figure(figsize=(6, 4))
plt.scatter(df_clean["math_score"], df_clean["science_score"])
plt.title("Math Score vs Science Score")
plt.xlabel("Math Score")
plt.ylabel("Science Score")
plt.show()

plt.figure(figsize=(6, 4))
plt.hist(df_clean["average_score"], bins=5)
plt.title("Distribution of Average Scores")
plt.xlabel("Average Score")
plt.ylabel("Number of Students")
plt.show()
```

### Exercise 6.1: Create One Chart and Explain It

```python
plt.figure(figsize=(6, 4))
plt.scatter(df_clean["english_score"], df_clean["average_score"])
plt.title("English Score vs Average Score")
plt.xlabel("English Score")
plt.ylabel("Average Score")
plt.show()

explanation = "Students with higher English scores also tend to have higher overall averages in this toy dataset."
print(explanation)
```

**TA Note:** Any correctly labeled plot using the data is acceptable.

**Hint to give:** Ask what variables belong on x and y axes.

**Common student mistakes:** Forgetting plt.show() or leaving out axis labels.
