## Section 1: What is Supervised Learning?

Supervised learning uses examples with known answers.

- `X`: features, or input columns.
- `y`: target, or answer column.
- Training data: examples used to fit the model.
- Test data: examples used to check performance.
- Classification predicts categories.
- Regression predicts numbers.

Examples: flower species classification, survival classification, and exam score regression.

## Section 2: Classification Dataset

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.datasets import load_iris

iris = load_iris()
X = pd.DataFrame(iris.data, columns=iris.feature_names)
y = pd.Series(iris.target, name="species")
iris_df = X.copy()
iris_df["species_name"] = y.map(dict(enumerate(iris.target_names)))
iris_df.head()
```

```python
print(X.shape)
print(iris.target_names)
print(iris_df["species_name"].value_counts())
```

## Section 3: Train/Test Split

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.25, random_state=42, stratify=y
)

print(X_train.shape, X_test.shape)
```

## Section 4: Model 1 - K-Nearest Neighbors

KNN predicts using nearby training examples.

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score

knn = KNeighborsClassifier(n_neighbors=3)
knn.fit(X_train, y_train)
knn_predictions = knn.predict(X_test)
knn_accuracy = accuracy_score(y_test, knn_predictions)
print("KNN accuracy:", knn_accuracy)
```

### Exercise 4.1: Try Different Values of k

```python
k_results = []
for k in [1, 3, 5, 7, 9]:
    model = KNeighborsClassifier(n_neighbors=k)
    model.fit(X_train, y_train)
    predictions = model.predict(X_test)
    k_results.append({"k": k, "accuracy": accuracy_score(y_test, predictions)})
pd.DataFrame(k_results)
```

**TA Note:** Accuracy may be around 0.9-1.0 depending on split and k.

**Hint to give:** Ask students what loop variable changes each time.

**Common student mistakes:** Training the model but forgetting to call predict on X_test.

## Section 5: Model 2 - Decision Tree

```python
from sklearn.tree import DecisionTreeClassifier

tree = DecisionTreeClassifier(max_depth=3, random_state=42)
tree.fit(X_train, y_train)
tree_predictions = tree.predict(X_test)
tree_accuracy = accuracy_score(y_test, tree_predictions)
print("Decision tree accuracy:", tree_accuracy)
```

### Exercise 5.1: Change max_depth and Compare Models

```python
depth_results = []
for depth in [1, 2, 3, 4, None]:
    model = DecisionTreeClassifier(max_depth=depth, random_state=42)
    model.fit(X_train, y_train)
    predictions = model.predict(X_test)
    depth_results.append({"max_depth": depth, "accuracy": accuracy_score(y_test, predictions)})
pd.DataFrame(depth_results)
```

**TA Note:** Depth 1 often underfits; deeper trees may fit better but can overfit.

**Hint to give:** Ask what max_depth controls in a tree.

**Common student mistakes:** Comparing models trained on different splits without noticing it.

## Section 6: Evaluation

Accuracy is useful, but it is not the whole story. Confusion matrices show which classes get mixed up.

```python
from sklearn.metrics import confusion_matrix, classification_report

print(confusion_matrix(y_test, knn_predictions))
print(classification_report(y_test, knn_predictions, target_names=iris.target_names))
```

### Exercise 6.1: Identify Confused Classes and Interpret

```python
report_sentence = "The model performs well overall. Most mistakes, if any, are between similar iris species rather than setosa."
print(report_sentence)
```

## Section 7: Regression Mini-Demo

Regression predicts a number. Here we predict exam score from study hours.

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

study_hours = np.array([1, 2, 3, 4, 5, 6, 7, 8]).reshape(-1, 1)
exam_scores = np.array([55, 60, 65, 70, 76, 80, 86, 90])

reg = LinearRegression()
reg.fit(study_hours, exam_scores)
predicted_scores = reg.predict(study_hours)
mse = mean_squared_error(exam_scores, predicted_scores)

print("Slope:", reg.coef_[0])
print("Intercept:", reg.intercept_)
print("MSE:", mse)

plt.figure(figsize=(6, 4))
plt.scatter(study_hours, exam_scores, label="Actual")
plt.plot(study_hours, predicted_scores, color="red", label="Predicted line")
plt.title("Study Hours vs Exam Score")
plt.xlabel("Study Hours")
plt.ylabel("Exam Score")
plt.legend()
plt.show()
```

### Exercise 7.1: Predict a New Score and Interpret the Slope

```python
new_hours = np.array([[9]])
predicted_new_score = reg.predict(new_hours)[0]
print(predicted_new_score)
print("The slope means each additional study hour is associated with about this many more exam points in the toy data:", reg.coef_[0])
```

## Section 9: Challenge

Compare KNN, decision tree, and logistic regression.

Output:

- small results table
- short paragraph interpreting results

```python
from sklearn.linear_model import LogisticRegression

models = [
    ("KNN", KNeighborsClassifier(n_neighbors=3)),
    ("Decision Tree", DecisionTreeClassifier(max_depth=3, random_state=42)),
    ("Logistic Regression", LogisticRegression(max_iter=200, random_state=42)),
]

challenge_results = []
for name, model in models:
    model.fit(X_train, y_train)
    predictions = model.predict(X_test)
    challenge_results.append({"model": name, "accuracy": accuracy_score(y_test, predictions)})

results_table = pd.DataFrame(challenge_results)
print(results_table)
print("Interpretation: Several simple models perform well on Iris because the classes are fairly separable.")
```

**TA Note:** Students may see all three models perform well.

**Hint to give:** Ask them to make a table with the same columns for each model.

**Common student mistakes:** Only reporting the best model without mentioning test size or limitations.
