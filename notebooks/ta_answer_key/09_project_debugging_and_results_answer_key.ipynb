## Section 1: Common Import Errors

```python
# Broken:
# import panda as pd

import pandas as pd
print("pandas imported")
```

## Section 2: File Path Errors

```python
from pathlib import Path

DATA_DIR = Path("../../data")
if not (DATA_DIR / "sample_students.csv").exists():
    DATA_DIR = Path("data")

csv_path = DATA_DIR / "sample_students.csv"
print("Using:", csv_path)
```

```python
df = pd.read_csv(csv_path)
print(df.head())
```

## Section 3: Missing-value Errors

```python
df_fixed = df.copy()
df_fixed["english_score"] = df_fixed["english_score"].fillna(df_fixed["english_score"].mean())
print(df_fixed.isna().sum())
```

## Section 4: Model Training Errors

```python
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score

X = df_fixed[["math_score", "english_score", "science_score"]]
y = (df_fixed["average_score"] if "average_score" in df_fixed.columns else X.mean(axis=1)) >= 85
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
model = KNeighborsClassifier(n_neighbors=3)
model.fit(X_train, y_train)
predictions = model.predict(X_test)
print("Accuracy:", accuracy_score(y_test, predictions))
```

## Section 5: Plotting Errors

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(6, 4))
plt.scatter(df_fixed["math_score"], df_fixed["science_score"])
plt.title("Math vs Science Scores")
plt.xlabel("Math Score")
plt.ylabel("Science Score")
plt.show()
```

## Section 6: How to Create a Results Table

```python
results_table = pd.DataFrame([
    {"Model": "KNN", "Metric": "accuracy", "Value": 0.80, "Notes": "Toy example"},
    {"Model": "Decision Tree", "Metric": "accuracy", "Value": 0.75, "Notes": "Toy example"},
])
results_table
```

## Section 7: How to Interpret Results

```python
interpretation = "In this toy example, KNN has slightly higher accuracy than the decision tree. Because the dataset is tiny, this result should be treated as practice, not a final scientific conclusion."
print(interpretation)
```

## Section 8: How to Write Limitations

```python
limitations = [
    "The dataset is small and synthetic.",
    "The train/test split is small, so accuracy is unstable.",
    "The features may not capture all important causes.",
]
for limitation in limitations:
    print("-", limitation)
```
