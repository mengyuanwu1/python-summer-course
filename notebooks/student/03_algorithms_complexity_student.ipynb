## Section 2: Linear Search

Linear search checks items one at a time.

```python
def linear_search(items, target):
    for index, item in enumerate(items):
        if item == target:
            return index
    return -1

print(linear_search([3, 8, 2, 7], 2))
print(linear_search([3, 8, 2, 7], 5))
```

## Section 3: Binary Search

Binary search works only on sorted lists. It repeatedly checks the middle and removes half the remaining possibilities.

```python
def binary_search(sorted_items, target):
    low = 0
    high = len(sorted_items) - 1

    while low <= high:
        mid = (low + high) // 2
        if sorted_items[mid] == target:
            return mid
        elif sorted_items[mid] < target:
            low = mid + 1
        else:
            high = mid - 1
    return -1

print(binary_search([1, 3, 5, 7, 9, 11], 7))
```

## Section 4: Recursion

A recursive function calls itself. It needs a base case and a recursive case.

```python
def factorial(n):
    if n == 0:
        return 1
    return n * factorial(n - 1)

def countdown(n):
    if n == 0:
        print("done")
    else:
        print(n)
        countdown(n - 1)

def recursive_sum(numbers):
    if len(numbers) == 0:
        return 0
    return numbers[0] + recursive_sum(numbers[1:])

print(factorial(5))
countdown(3)
print(recursive_sum([1, 2, 3, 4]))
```

## Section 5: Sorting

Sorting makes data easier to search, compare, and present.

```python
def selection_sort(numbers):
    numbers = numbers.copy()
    for i in range(len(numbers)):
        min_index = i
        for j in range(i + 1, len(numbers)):
            if numbers[j] < numbers[min_index]:
                min_index = j
        numbers[i], numbers[min_index] = numbers[min_index], numbers[i]
    return numbers

print(selection_sort([5, 2, 9, 1]))
```

## Section 6: Timing with `timeit`

Timing experiments help us see runtime differences. Results vary by computer, so compare patterns, not exact numbers.

```python
import timeit

for size in [100, 1000, 10000]:
    data = list(range(size))
    target = size - 1
    elapsed = timeit.timeit(lambda: linear_search(data, target), number=100)
    print(size, round(elapsed, 5))

sample = list(range(200, 0, -1))
print("selection sort:", round(timeit.timeit(lambda: selection_sort(sample), number=10), 5))
print("built-in sorted:", round(timeit.timeit(lambda: sorted(sample), number=10), 5))
```

## Section 7: Complexity in Plain Language

- Constant time: does not grow much with input size, such as list indexing.
- Linear time: grows like the number of items, such as linear search.
- Quadratic time: nested loops over the same data, such as selection sort.
- Logarithmic time: repeatedly cuts the problem down, such as binary search.
