## Section 1: What is Unsupervised Learning?

Unsupervised learning looks for patterns when we do not give the model labels. Examples include customer groups, text topics, and image groups.

## Section 2: Load Dataset

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.datasets import load_iris

iris = load_iris()
X = pd.DataFrame(iris.data, columns=iris.feature_names)
hidden_labels = iris.target
X.head()
```

## Section 3: Visualize Before Clustering

```python
plt.figure(figsize=(6, 4))
plt.scatter(X["sepal length (cm)"], X["petal length (cm)"])
plt.title("Iris Features Before Clustering")
plt.xlabel("Sepal Length (cm)")
plt.ylabel("Petal Length (cm)")
plt.show()
```

### Exercise 3.1: Write One Observation

```python
observation = "The scatter plot suggests some points form a visible group, but some groups overlap."
print(observation)
```

## Section 4: Standardize Features

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
print(X_scaled[:3])
```

## Section 5: K-means Clustering

```python
from sklearn.cluster import KMeans

cluster_results = {}
for k in [2, 3, 4]:
    kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
    cluster_results[k] = kmeans.fit_predict(X_scaled)

    plt.figure(figsize=(6, 4))
    plt.scatter(X["sepal length (cm)"], X["petal length (cm)"], c=cluster_results[k])
    plt.title(f"K-means Clusters with k={k}")
    plt.xlabel("Sepal Length (cm)")
    plt.ylabel("Petal Length (cm)")
    plt.show()
```

### Exercise 5.1: Compare k=2, k=3, and k=4

```python
k_choice_sentence = "k=3 seems reasonable because Iris has three known species, but the model did not see those labels."
print(k_choice_sentence)
```

**TA Note:** Students should compare visual separation, not just memorize that Iris has 3 classes.

**Hint to give:** Ask what evidence appears in the plots.

**Common student mistakes:** Claiming clusters prove true species without checking labels or limitations.

## Section 6: Choosing K

```python
inertias = []
k_values = range(1, 8)
for k in k_values:
    model = KMeans(n_clusters=k, random_state=42, n_init=10)
    model.fit(X_scaled)
    inertias.append(model.inertia_)

plt.figure(figsize=(6, 4))
plt.plot(list(k_values), inertias, marker="o")
plt.title("Elbow Method for K-means")
plt.xlabel("Number of Clusters k")
plt.ylabel("Inertia")
plt.show()
```

### Exercise 6.1: Is the Elbow Obvious?

```python
elbow_sentence = "The elbow is not perfectly obvious, which shows that choosing k often requires judgment."
print(elbow_sentence)
```

## Section 7: PCA for Visualization

```python
from sklearn.decomposition import PCA

pca = PCA(n_components=2, random_state=42)
X_pca = pca.fit_transform(X_scaled)
labels_k3 = cluster_results[3]

plt.figure(figsize=(6, 4))
plt.scatter(X_pca[:, 0], X_pca[:, 1], c=labels_k3)
plt.title("K-means Clusters Visualized with PCA")
plt.xlabel("Principal Component 1")
plt.ylabel("Principal Component 2")
plt.show()
```
