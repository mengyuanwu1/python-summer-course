## Section 1: 5-minute Presentation Structure

Recommended structure:

2. Dataset or code source
3. Method
4. Results
5. Limitations and conclusion

## Section 3: How to Explain Code Clearly

```python
code_explanation = "I first cleaned the data, then standardized numeric features, then fit K-means. The model assigned each customer to a cluster, which I visualized with PCA."
print(code_explanation)
```

## Section 4: How to Explain Results Clearly

```python
result_explanation = "The clusters suggest three customer groups with different spending and visit patterns. This supports the idea that unsupervised learning can help summarize customer behavior, but the labels are exploratory."
print(result_explanation)
```

## Section 5: Final Report Structure

```python
report_outline = pd.DataFrame([
    {"Section": "Title and Abstract", "Contents": "Short title and 3-5 sentence summary"},
    {"Section": "Introduction / WHY", "Contents": "Motivation and at least five scholarly sources"},
    {"Section": "Data / WHAT", "Contents": "Dataset, features, source, limitations"},
    {"Section": "Methods / HOW", "Contents": "Code workflow and model choices"},
    {"Section": "Results", "Contents": "Figures, tables, and interpretation"},
    {"Section": "Limitations and Conclusion", "Contents": "What the project shows and cannot show"},
    {"Section": "References", "Contents": "All cited papers, datasets, and tutorials"},
])
report_outline
```

## Debugging or Common Mistakes Section

- Trying to explain too many results in 5 minutes.
- Showing code screenshots instead of explaining the workflow.
- Missing WHY or citations in the report.
- Making figures too small to read.
- Forgetting limitations.
